```python
import math
import matplotlib.pyplot as plt
import numpy as np
import math

def sigmoid(x):
  return 1+(1 / (1 + math.exp((-x+45)/12)))
```

```python
x=np.arange(0,90)
n=10
y1=[n for x in x]
y2=[2*n*sigmoid(x) for x in x]
y3=[3*n*sigmoid(x) for x in x]
plt.plot(x,y1, label = "NEN Holders(No Staking)")
plt.plot(x,y2, label = "Single Asset Stakers")
plt.plot(x,y3, label = "LP Token Stakers")
plt.xlabel("d")
plt.ylabel("P")
plt.legend()
plt.show()
```
